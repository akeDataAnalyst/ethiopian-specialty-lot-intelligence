#### Create Mock Data (Realistic Ethiopian Lots)

```python
np.random.seed(42)

regions = ['Yirgacheffe', 'Sidama', 'Guji', 'Limmu', 'Kaffa']
districts = ['Yirga Chefe', 'Gedeb', 'Shakiso', 'Hambela', 'Bensa', 'Uraga']
suppliers = ['Daye Bensa', 'Hambela Coffee', 'Misty Valley', 'Yirgacheffe Coffee Farmers', 
             'Guji Highland', 'Sidama Crown', 'Kaffa Organic Union', 'Red Cherry Export']

n = 50

data = {
    'lot_id': [f"ETH-2025-{reg[:3].upper()}-{str(i).zfill(3)}" for i, reg in enumerate(np.random.choice(regions, n), 1)],
    'supplier_name': np.random.choice(suppliers, n),
    'washing_station': np.random.choice(['Hambela', 'Daye Bensa', 'Gedeb', 'Uraga', 'Shakiso', 'Bensa'], n),
    'region': np.random.choice(regions, n),
    'district': np.random.choice(districts, n),
    'processing_method': np.random.choice(['Washed', 'Natural', 'Honey'], n, p=[0.55, 0.35, 0.10]),
    'harvest_year': 2025,
    'harvest_period': np.random.choice(['Oct-Nov 2025', 'Nov-Dec 2025', 'Dec-Jan 2026'], n),
    'arrival_date': pd.date_range(start='2026-02-01', periods=n, freq='3D'),  # Realistic arrival window
    'grade_ecx': np.random.choice(['Grade 1', 'Grade 2'], n, p=[0.68, 0.32]),
    'defects_per_300g': np.random.randint(0, 14, n),
    'moisture_pct': np.round(np.random.uniform(9.8, 12.1, n), 2),
    'water_activity': np.round(np.random.uniform(0.58, 0.67, n), 3),
    'sca_score': np.round(np.random.uniform(82.0, 91.5, n), 1),
    'flavor_profile': np.random.choice([
        'Jasmine, Bergamot, Citrus', 'Blueberry, Chocolate, Bright Acidity',
        'Floral, Peach, Tea-like', 'Strawberry, Tropical Fruit, Winey',
        'Lemon, Honey, Complex', 'Black Tea, Red Berry, Silky'
    ], n),
    'quantity_bags': np.random.randint(12, 220, n),
    'price_per_kg_usd': np.round(np.random.uniform(5.2, 9.8, n), 2),
    'traceability_level': np.random.choice(['Full', 'Washing Station', 'Exporter'], n, p=[0.45, 0.35, 0.20]),
    'sustainability_cert': np.random.choice(['Organic', 'Volcafe Verified', 'Rainforest Alliance', 'None'], n),
    'status': np.random.choice(['Available', 'Reserved'], n, p=[0.85, 0.15])
}

df = pd.DataFrame(data)

df['available_quantity_kg'] = df['quantity_bags'] * 60.0
df['last_updated'] = datetime.now()
df['is_specialty'] = df['sca_score'] >= 80.0

print("Realistic Mock Dataset Created")
print(f"Total Lots: {len(df)}")
print(f"Specialty Lots (80+ SCA): {df['is_specialty'].sum()}")
print(f"Harvest Year: {df['harvest_year'].unique()[0]}")
```
